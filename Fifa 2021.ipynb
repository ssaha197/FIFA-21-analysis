```python
data['pace'].fillna(data['pace'].mean(), inplace = True)
data['shooting'].fillna(data['shooting'].mean(), inplace = True)
data['passing'].fillna(data['passing'].mean(), inplace = True)
data['dribbling'].fillna(data['dribbling'].mean(), inplace = True)
data['defending'].fillna(data['defending'], inplace = True)
data['physic'].fillna(data['physic'].mean(), inplace = True)
data['attacking_crossing'].fillna(data['attacking_crossing'].mean(), inplace = True)
data['attacking_finishing'].fillna(data['attacking_finishing'].mean(), inplace = True)
data['attacking_heading_accuracy'].fillna(data['attacking_heading_accuracy'].mean(), inplace = True)
data['attacking_short_passing'].fillna(data['attacking_short_passing'].mean(), inplace = True)
data['attacking_volleys'].fillna(data['attacking_volleys'].mean(), inplace = True)
data['skill_dribbling'].fillna(data['skill_dribbling'].mean(), inplace = True)
data['skill_curve'].fillna(data['skill_curve'].mean(), inplace = True)
data['skill_fk_accuracy'].fillna(data['skill_fk_accuracy'].mean(), inplace = True)
data['skill_long_passing'].fillna(data['skill_long_passing'].mean(), inplace = True)
data['skill_ball_control'].fillna(data['skill_ball_control'].mean(), inplace = True)
data['movement_acceleration'].fillna(data['movement_acceleration'].mean(), inplace = True)
data['movement_sprint_speed'].fillna(data['movement_sprint_speed'].mean(), inplace = True)
data['movement_agility'].fillna(data['movement_agility'].mean(), inplace = True)
data['movement_reactions'].fillna(data['movement_reactions'].mean(), inplace = True)
data['movement_balance'].fillna(data['movement_balance'].mean(), inplace = True)
data['power_shot_power'].fillna(data['power_shot_power'].mean(), inplace = True)
data['power_jumping'].fillna(data['power_jumping'].mean(), inplace = True)
data['power_stamina'].fillna(data['power_stamina'].mean(), inplace = True)
data['power_strength'].fillna(data['power_strength'].mean(), inplace = True)
data['power_long_shots'].fillna(data['power_long_shots'].mean(), inplace = True)
data['mentality_aggression'].fillna(data['mentality_aggression'].mean(), inplace = True)
data['mentality_interceptions'].fillna(data['mentality_interceptions'].mean(), inplace = True)
data['mentality_positioning'].fillna(data['mentality_positioning'].mean(), inplace = True)
data['mentality_vision'].fillna(data['mentality_vision'].mean(), inplace = True)
data['mentality_penalties'].fillna(data['mentality_penalties'].mean(), inplace = True)
data['mentality_composure'].fillna(data['mentality_composure'].mean(), inplace = True)
data['defending_marking'].fillna(data['defending_marking'].mean(), inplace = True)
data['defending_standing_tackle'].fillna(data['defending_standing_tackle'].mean(), inplace = True)
data['defending_sliding_tackle'].fillna(data['defending_sliding_tackle'].mean(), inplace = True)
data['goalkeeping_diving'].fillna(data['goalkeeping_diving'].mean(), inplace = True)
data['goalkeeping_handling'].fillna(data['goalkeeping_handling'].mean(), inplace = True)
data['goalkeeping_kicking'].fillna(data['goalkeeping_kicking'].mean(), inplace = True)
data['goalkeeping_positioning'].fillna(data['goalkeeping_positioning'].mean(), inplace = True)
data['goalkeeping_reflexes'].fillna(data['goalkeeping_reflexes'].mean(), inplace = True)
data['weight_kg'].fillna('70', inplace = True)
data['contract_valid_until'].fillna(2021, inplace = True)
data['height_cm'].fillna('172', inplace = True)
data['loaned_from'].fillna('None', inplace = True)
data['joined'].fillna('10-01-2021', inplace = True)
data['team_jersey_number'].fillna(19, inplace = True)
data['body_type'].fillna('Normal', inplace = True)
data['player_positions'].fillna('CM', inplace = True)
data['club_name'].fillna('No Club', inplace = True)
data['work_rate'].fillna('Medium/Medium', inplace = True)
data['skill_moves'].fillna(data['skill_moves'].median(), inplace = True)
data['weak_foot'].fillna(3, inplace = True)
data['preferred_foot'].fillna('Right', inplace = True)
data['international_reputation'].fillna(1, inplace = True)
data['wage_eur'].fillna('200000', inplace = True)
data['dob'].fillna('11-09-2001', inplace = True)
data['age'].fillna('19', inplace = True)
data['league_rank'].fillna(data['league_rank'].median(), inplace = True)
data.fillna(0, inplace = True)
```

```python
def general(data):
    return int(round((data[['pace', 'shooting', 'passing', 'dribbling','defending','physic']].mean()).mean()))

def attack(data):
    return int(round((data[['attacking_crossing','attacking_finishing','attacking_heading_accuracy','attacking_short_passing','attacking_volleys']].mean()).mean()))

def skill(data):
    return int(round((data[['skill_dribbling','skill_curve','skill_fk_accuracy','skill_long_passing','skill_ball_control']].mean()).mean()))

def movement(data):
    return int(round((data[['movement_acceleration', 'movement_sprint_speed', 'movement_agility','movement_reactions','movement_balance']].mean()).mean()))

def power(data):
    return int(round((data[['power_shot_power', 'power_jumping', 'power_stamina', 'power_strength','power_long_shots']].mean()).mean()))

def mentality(data):
    return int(round((data[['mentality_aggression', 'mentality_interceptions', 'mentality_positioning', 'mentality_vision','mentality_penalties','mentality_composure']].mean()).mean()))

def defending(data):
    return int(round((data[['defending_marking', 'defending_standing_tackle', 'defending_sliding_tackle']].mean()).mean()))

def goalkeeping(data):
    return int(round((data[['goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes']].mean()).mean()))

def rating(data):
    return int(round((data[['overall', 'potential']].mean()).mean()))

def shooting(data):
    return int(round((data[['attacking_finishing', 'attacking_volleys', 'skill_fk_accuracy', 'power_shot_power','power_long_shots', 'mentality_penalties']].mean()).mean()))
```

```python
data['general'] = data.apply(general, axis = 1)
data['attack'] = data.apply(attack, axis = 1)
data['skill'] = data.apply(skill, axis = 1)
data['movement'] = data.apply(movement, axis = 1)
data['power'] = data.apply(power, axis = 1)
data['mentality'] = data.apply(mentality, axis = 1)
data['defending'] = data.apply(defending, axis = 1)
data['goalkeeping'] = data.apply(goalkeeping, axis = 1)
data['rating'] = data.apply(rating, axis = 1)
data['shooting'] = data.apply(shooting, axis = 1)
```
